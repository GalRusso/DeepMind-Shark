```python
environment = dbutils.widgets.get("environment")
board_id = dbutils.widgets.get("board_id")
group_id = dbutils.widgets.get("group_id")
vertical = dbutils.widgets.get("vertical")
data_domain = dbutils.widgets.get("data_domain")
file_name_prefix = dbutils.widgets.get("file_name_prefix")
```

```python
json_data = monday.get_board_items(
    api_key=dbutils.secrets.get(scope="hoangl", key="monday_api_key"),
    board_id=board_id,
    group_id=group_id
)
```

```python
from datetime import datetime
import json

# Set the S3 path with file_name containing full datetime
now = datetime.now()
data_source = "monday"

# Create directory path and file path separately
dir_path = f"/Volumes/af_delivery_{environment}/data_collection/{vertical}/{data_domain}/{data_source}/{now:%Y}/{now:%m}/{now:%d}"
file_name = f"{file_name_prefix}_{now:%Y%m%d_%H%M%S}.json"

volume_path = f"{dir_path}/{file_name}"

# Create directory if it doesn't exist
dbutils.fs.mkdirs(dir_path)

# Convert data to JSON string
json_content = json.dumps(json_data, indent=2)

# Write to S3 using dbutils.fs.put
dbutils.fs.put(volume_path, json_content, overwrite=True)

print(f"Successfully wrote {len(json_data)} items to {volume_path}")
```
